```python
from matplotlib import pyplot as plt
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
x_values=[1,4,-1,0,-2]
y_values=[2,7,3,-1,-3]
colors=['r','g','y','b','b']
shapes=['o','*','s','^','>']
ax.set_xlabel('X AXIS')
ax.set_ylabel('Y AXIS')
#ax.scatter(x,y,color=colors,marker=shapes)
for i in range(len(x_values)):
    ax.scatter(x_values[i],y_values[i],color=colors[i],marker=shapes[i])

ax.set_title('Scatter plot')

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
letter = ['A','B','C','D','E','F']
num = [7,3,12,4,2,1]
ax.bar(letter,num)
plt.show()
```
